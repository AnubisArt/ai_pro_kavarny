```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.read_csv('coffee-listings-from-all-walmart-stores.csv')
data

X = data.drop(columns = ['title','coffee_type','reviews','seller_name','thumbnail','weight'])
#input data is (coffee type TC), rating, and price in $
y = data['title']
cft = data["coffee_type"]

model = DecisionTreeClassifier()
model.fit(X,y)
predictions = model.predict([[3.8, 13.92,1000]])
predictions

cft
```
